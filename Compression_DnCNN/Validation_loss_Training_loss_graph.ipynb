```python
# Google Colab 환경에서 진행
import matplotlib.pyplot as plt
import numpy as np
```

```python
train1 = np.loadtxt('drive/MyDrive/Colab Notebooks/Test/train_loss_results.txt')
train2 = np.loadtxt('drive/MyDrive/Colab Notebooks/Test/train_loss_results (1).txt')
train3 = np.loadtxt('drive/MyDrive/Colab Notebooks/Test/train_loss_results (2).txt')
```

```python
valid1 = np.loadtxt('drive/MyDrive/Colab Notebooks/Test/valid_loss_results.txt')
valid2 = np.loadtxt('drive/MyDrive/Colab Notebooks/Test/valid_loss_results (1).txt')
valid3 = np.loadtxt('drive/MyDrive/Colab Notebooks/Test/valid_loss_results (2).txt')
```

```python
train = np.concatenate((train1, train2, train3), axis=0)
valid = np.concatenate((valid1, valid2, valid3), axis=0)
```

```python
x = np.arange(0, len(valid))

fig, ax1 = plt.subplots()

ax1.set_xlabel('epoch')
ax1.set_ylabel('Validation Loss')

line1 = ax1.plot(x, valid, color = 'Blue', label = 'valid_loss')

ax2 = ax1.twinx()
ax2.set_ylabel('Training Loss')

line2 = ax2.plot(x, train, color = 'Orange', label = 'train_loss')

lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper right')

plt.title('Loss')
plt.grid()

plt.show()
```
